<h1>Compressed Sensing 101</h1>
Compressed sensing (CS) is an emerging field of math and engineering that challenges the conventional paradigms of data acquisition. Since the first publication(Candes, 2007?), the field has grown to over xxx academic papers and has provided the foundation for novel innovation in medical imaging, atronomy, and digital photography etc..(find references). 

Many signals have an inherent low-dimensional structure, meaning all the information can be contained in relatively few coefficients in a transformed domain. In other words, there exists a domain where the signal is sparse. JPEG and mp3 compression algorithms make use of this to greatly reduce the size of media files. Knowing that signal information can be reduced to a few coefficients, CS asks whether we can exploit this directly at the sampling stage.

highly sampled data -> compression
sparsely sampled data

Generally, CS works as follows:
1. The signal has a sparse representation in some domain.
2. We undersample the data.
3. Reconstruct the full signal by promoting sparsity.

This tutorial applies CS to a simple 1D signal:

$x(t) = 10Sin(2\pi f_1 t + \phi_1) + 20Sin(2\pi f_2t + \phi_2)$

```python
f1 = 10. #[Hz]
f2 = 40. #[Hz]
phi1 = pi/5.
phi2 = pi/9.
duration = 0.5 #[s]

xt, t = demo.test_signal(f1,f2,phi1,phi2, duration)
```

<h1>Sampling</h1>
<p>
Knowing the signal can be represented by only a few cofficients, we can now try to exploit that directly by taking fewer samples.
</p>

<b>Uniform sampling</b>
<p>Using uniform sampling we are limited by the Nyquist rate. We do not have enough samples to uniquely define our signal and we begin to see aliased coefficients in the Fourier domain.</p>

```python
sample_slider = widgets.IntSliderWidget(min=20, max=xt.size/4, step=5, value=xt.size/2)
w=widgets.interactive(demo.uniform_sample,n_samples=sample_slider)
display(w)
```

<b>Aliasing</b>
<p>Lets take a closer look aliasing. The cause of aliasing is that multiple signals can have the same value at each sample.</p>

```python
    from numpy.random import random_sample    
    ny = (2000.)/2
    t = arange(0, duration, 1/(2*ny))
    sig1 = sin(2*pi*10*t)
    sig2 = sin(2*pi*(10 + ny/8)*t)
    plt.plot(t, sig1);plt.xlim(0,.1)
    plt.plot(t, sig2, 'g')
    samples = random_sample(duration*(ny/8))*duration
    plt.plot(samples,sin(2*pi*10*samples), 'bs')
    plt.plot(samples, sin(2*pi*(10 + ny/8)*samples), 'gs')
    ax = plt.gca()
    ax.set_xticks(samples, minor=True)
    ax.xaxis.grid(which='minor')
```
