<h1>Compressed Sensing 101</h1>
Compressed sensing (CS) is an emerging field of math and engineering that challenges the conventional paradigms of data acquisition. Since the first publication(Candes, 2007?), the field has grown to over xxx academic papers and has provided the foundation for novel innovation in medical imaging, atronomy, and digital photography etc..(find references). 

Many signals have an inherent low-dimensional structure, meaning all the information is contained in relatively few coefficients in some transform domain. For example, harmonic signals reduce to a few Fourier coefficients, or an image can be compressed using a Wavelet transform. JPEG and mp3 compression algorithms rely on this to greatly reduce the size of media files. If we know that a signal has a sparse representation, CS asks whether we can exploit this directly by taking less samples during acquisition.

This tutorial applies the principles of CS to sampling the simple 1D signal:

$x(t) = 10Sin(2\pi f_1 t + \phi_1) + 20Sin(2\pi f_2t + \phi_2)$


```python
f1 = 10. #[Hz]
f2 = 40. #[Hz]
phi1 = pi/5.
phi2 = pi/9.
duration = 2.0#[s]

xt, t = demo.test_signal(f1,f2,phi1,phi2, duration)
```

<h1>Sparse Representation</h1>
<p>
 In this simple test case, we know the signal is sparse in the Fourier domain. All of the signal information is contained in just four complex Fourier coefficients.
</p>


```python
demo.ts_plot(t, xt)
```

```python
Xf,f  = demo.fft(xt, t)

demo.fourier_plot(f,Xf)
```

<h1>Sampling</h1>
<p>
If we know that our signal can be reduced to just a few coefficients, it seems incredibly inefficient to collect many times more samples during acquisition. Let's explore some sub-sampling strategies and see if we can reduce the amount of data we collect.
</p>

<b>Uniform sampling</b>
<p>
Uniform sampling has been the engineered acquisition strategy for decades. Shannon-Nyquist theory says we can uniquely represent our signal by sampling at least twice highest frequency. For our test signal that amounts > 40 samples. Considering it can be reduced to only 4 samples of information, we are grossly oversampling. Let's see what happens when try to collect less.
</p>

```python
sample_slider = widgets.IntSliderWidget(min=10, max=20, step=1, value=10)
w=widgets.interactive(demo.uniform_sample,ds_factor=sample_slider)
sample_slider.msg_throttle = 1
display(w)
```

<b>Aliasing</b>
<p>Lets take a closer look aliasing. The cause of aliasing is that multiple signals can have the same value at each sample.
</p>

```python
demo.aliased()
        
```

The problem is that our sampling pattern is periodic, and will be coherent with many period structures in the signal. Let's break the periodicity by sampling randomly and see what happens.

```python
  demo.randomized()
```

<p>The signals are now distinguishable when sampled</p>

<b>Random sampling</b>
<p>As shown above, we can avoid aliasing by using a random sampling pattern. Let's return to our test signal use this information to what happens when we use random sampling.</p>



```python
def random_wrapper(ds_factor):
    demo.random_sample(ds_factor)

sample_slider = widgets.IntSliderWidget(min=10, max=20, step=1, value=10)
w=widgets.interactive(random_wrapper,ds_factor=sample_slider)
display(w)
```

<p>
Note the differences in the Fourier spectrum compared to the aliased result we saw earlier. Since we broke the periodicity of our sampling pattern, aliased signals now become random noise. We still don't exactly recover our signal, but now that we have removed the ambiguity of aliasing we can consider ways to reconstruct our signal.
</p>


<h1>
Recovery
</h1>

In the previous section we showed that random sub-sampling turns aliased signals into random noise. All of these methods require some basic <i>a priori</i> information about our signal. Lets look at some possible methods to recover our full signal.

<b>
Thresholding
</b>
<p>
As a simple first step, lets start by thresholding. Here we assume we know something about the amplitude of the signal coefficients.
</p>

```python
def random_wrapper(threshold):
    demo.random_sample(threshold=threshold)

threshold_slider = widgets.IntSliderWidget(min=0, max=40000, step=1000, value=1000)
w=widgets.interactive(random_wrapper, threshold=threshold_slider)
display(w)
```

<b>
N-largest
</b>
<p>In this case we might know before hand how many sinusoids are in our signal. Lets try keeping only the n-largest coefficients.</p>

```python
def random_wrapper(ncoeffs):
    demo.random_sample(ncoeffs=ncoeffs)
    coeff_slider = widgets.IntSliderWidget(min=1, max=30, step=1, value=10)
w=widgets.interactive(random_wrapper, ncoeffs=coeff_slider)
display(w)
```

<b>
Optimization
</b>

The last method for recovery is perhaps the most robust but also the most computationally expensive. We can pose our signal recovery as an optimization problem, where we fit our data where we have actually sub-sampled using the sparsest Fourier coefficients we can.
